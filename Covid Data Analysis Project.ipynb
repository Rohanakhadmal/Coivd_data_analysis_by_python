```python
covid_df.drop(["Sno", "Time", "ConfirmedIndianNational", "ConfirmedForeignNational"], inplace = True, axis = 1)

# Now we dropping the columns that we didn't use.
```

```python
covid_df.head(5)
```

```python
covid_df['Date'] = pd.to_datetime(covid_df['Date'], format = '%Y-%m-%d')

#  Changing the date time format
```

```python
covid_df.head()
```

```python
# Now we have to find active_cases 
# the formula to find active_cases = Confirmed = (Cure + Deaths)
# Active cases are those patients who are not cured yet


covid_df['Actived_Cases'] = covid_df['Confirmed'] - (covid_df['Cured'] + covid_df['Deaths'])
```

```python
covid_df.tail(10)
```

```python
statewise = pd.pivot_table(covid_df, values = ['Confirmed', 'Deaths', 'Cured'], 
                           index = 'State/UnionTerritory', aggfunc = max)
```

```python
statewise['Recovery Rate'] = statewise['Cured'] * 100 / statewise['Confirmed']
```

```python
statewise['Mortality Rate'] = statewise['Deaths'] * 100 / statewise['Confirmed']
```

```python
statwise = statewise.sort_values(by = 'Confirmed', ascending = True)
```

```python
statewise.style.background_gradient(cmap = "cubehelix")


# So here we make pivot table that show recovery rate and mortality rate in percentage
# Pivot Table
```

```python
top_active_cases = covid_df.groupby(by = 'State/UnionTerritory').max()[['Actived_Cases', 'Date']].sort_values(by = ['Actived_Cases'], ascending = False).reset_index()
fig = plt.figure(figsize = (16, 9))
plt.title("Top 10 states with most active cases in India", size = 15)
ax = sns.barplot(data = top_active_cases.iloc[:10], y = "Actived_Cases", x = "State/UnionTerritory", linewidth = 2, edgecolor = 'black')
plt.show()


# So here's the barplot which shows top 10 staes with most active case in india.
```

```python
#  Top states with highest deaths

top_10_deaths = covid_df.groupby(by = 'State/UnionTerritory').max()[['Deaths', 'Date']].sort_values(by = ['Deaths'], ascending = False).reset_index()

fig = plt.figure(figsize = (18,5))
plt.title("Top states with most deaths", size = 25)
ax = sns.barplot(data = top_10_deaths.iloc[:15], y = "Deaths", x = 'State/UnionTerritory', linewidth = 2, edgecolor = 'black')
plt.xlabel('States')
plt.ylabel('Total Death Cases')
plt.show()
```

```python
# Growth Trend in Top 5 states 

fig = plt.figure(figsize=(10, 6))
ax = sns.lineplot(data = covid_df[covid_df['State/UnionTerritory'].isin(['Maharashtra', 'Karnataka', 'Kerala', 'Tamil Nadu', 'Uttar Pradesh'])], x='Date', y='Actived_Cases', hue='State/UnionTerritory')
ax.set_title("Top 5 Affected states in India", size=16)

```
